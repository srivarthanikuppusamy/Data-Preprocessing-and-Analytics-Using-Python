
Load a dataset and perform indexing, sorting, and basic filtering operations.


Import libraries

```python
import pandas as pd
import numpy as np
```

Create dataset

```python
arr={"index":[0,1,2],
     "name":["alice","bob","charlie"],
     "age":[25,30,35],
     "salary":[50000,60000,70000]}
```

Convert to DataFrame

```python
df = pd.DataFrame(arr)
df
```

Indexing

```python
index =df.set_index("age")
index
```

Sorting

```python
sort=df.sort_values("name")
sort
```

Filtering

```python
filter=df[df["salary"] >= 60000]
filter
```
