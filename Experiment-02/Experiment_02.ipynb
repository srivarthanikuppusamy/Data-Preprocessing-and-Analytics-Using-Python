Create a function to determine if a given number is prime. Use it inside a loop to
display all prime numbers between 1 and 100.


Function definition and Handling numbers ≤ 1

```python
def is_prime(num):
  if num <= 1:
       return False                       #Checking divisibility efficiently
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
            return False
  return True
                                            #Checking factor condition
print("Prime numbers between 1 and 100:")
for number in range(1, 101):
    if is_prime(number):
        print(number,end="  ")              #Printing prime numbers from 1 to 100
```
