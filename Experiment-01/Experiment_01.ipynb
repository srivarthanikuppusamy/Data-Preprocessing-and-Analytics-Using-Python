Write a Python program to compute the sum, average, and product of three user-input numbers.


Taking input from user

```python
num1 = int (input("Enter first number: "))
num2 = int (input("Enter second number: "))
num3 = int (input ("Enter third number: "))
```

Calculating sum

```python
sum = num1 + num2 + num3
```

Calculating average

```python
average = sum / 3
```

Calculating product

```python
product = num1 * num2 * num3
```

Displaying results

```python
print("Sum =", sum)
print("Average =", average)
print("Product =", product)
```
