
Detect and remove outliers from a dataset using the Z-score method.


Import libraries

```python
import numpy as np
from scipy import stats
```

Create dataset

```python
arr = [10, 12, 11, 13, 12, 100, 11, 12, 13, 11]
```

Convert to array

```python
data = np.array(arr)
data
```

Calculate Z-score

```python
z_scores = np.abs(stats.zscore(data))
z_scores
```

Detect outliers

```python
outliers=z_scores > 2.5
outliers
```

Remove outliers

```python
cleaned_data=data[~outliers]
cleaned_data
```
